```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
base = pd.read_excel('Investimento_x_Venda.xlsx')
```

```python
plt.Figure(figsize=(10, 6))

plt.scatter(base['Investimento em marketing'], base['Venda Qtd'], alpha=0.5)

# definindo o começo e o fim da linha
x0 = base['Investimento em marketing'][0]
y0 = base['Venda Qtd'][0]
x1 = base['Investimento em marketing'][6]
y1 = base['Venda Qtd'][6]

plt.plot([x0, x1], [y0, y1], markersize=10) 

plt.xlabel('Investimento em Marketing')
plt.ylabel('Venda Qtd')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(base[['Investimento em marketing']], base['Venda Qtd'])

#visualizando os coeficientes
print( modelo.coef_)
print(modelo.intercept_)
```

```python
plt.Figure(figsize=(10, 6))

plt.scatter(base['Investimento em marketing'], base['Venda Qtd'], alpha=0.5)

x = np.array(base['Investimento em marketing'])
y = modelo.intercept_ + (x * modelo.coef_)

plt.plot(x, y, "r")

plt.xlabel('Investimento em Marketing')
plt.ylabel('Venda Qtd')
plt.show() 
```

```python
# Qual o estoque da loja para o investimento de 75k

modelo.predict([[75]])

#Visualizar
plt.Figure(figsize=(10, 6))
plt.scatter(base['Investimento em marketing'], base['Venda Qtd'], alpha=0.5)
plt.scatter(75, modelo.predict([[75]]), color='red', s=100, label='Predição para 75k')

x = np.array(base['Investimento em marketing'])
y = modelo.intercept_ + (x * modelo.coef_)

plt.plot(x, y, "r")

plt.xlabel('Investimento em Marketing')
plt.ylabel('Venda Qtd')
plt.show() 
```
