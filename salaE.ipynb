```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
base = pd.read_excel('Investimento_x_Venda.xlsx')
```

```python
plt.Figure(figsize=(10, 6))
plt.scatter(base['Investimento em marketing'], base['Venda Qtd'], alpha=0.5)

# definindo o come√ßo e o fim da linha
x0 = base['Investimento em marketing'][0]
y0 = base['Venda Qtd'][0]
x1 = base['Investimento em marketing'][6]
y1 = base['Venda Qtd'][6]

plt.plot([x0, x1], [y0, y1], markersize=10) 

plt.xlabel('Investimento em Marketing')
plt.ylabel('Venda Qtd')
plt.show()
```
